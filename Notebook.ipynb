## Heat map: 128x64x18

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import sys
```

```python
MARKET_PATH = "./market_data/"
heatmap_path = "trainK/"
```

```python
heatmaps = os.listdir(os.path.join(MARKET_PATH, heatmap_path))
```

```python
for sample in heatmaps:
    s_heatmap = np.load(os.path.join(MARKET_PATH, heatmap_path, sample))
    plt.imshow(s_heatmap.sum(2), cmap='gray')
    plt.show()
    if(input("Press Enter to continue/ press 'q' to quit: ") == 'q'):
        break
```

## Alphapose sample

```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import sys
import copy
```

```python
AP_PATH = "../AlphaPose/examples/"
```

```python
img = plt.imread(os.path.join(AP_PATH, "demo/1.jpg"))
with open(os.path.join(AP_PATH, "alphapose-results.json"), 'r') as f:
    re = json.load(f)

sample = np.array(re[0]['keypoints']).reshape(-1, 3)
```

```python
for i in range(17): 
    img_tmp = copy.copy(img) 
    img_tmp[int(round(sample[i,1]))-3:int(round(sample[i, 1]))+3, int(round(sample[i, 0]))-3:int(round(sample[i, 0]))+3,:] = [255, 0, 0] 
    plt.imshow(img_tmp)
    plt.show()
    if(input("Press Enter to continue/ press 'q' to quit: ") == 'q'):
        break
```
